```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Read Image and Apply Adaptive Thresholding

```python
# Load the image
image = cv2.imread('./images/car.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply adaptive thresholding
thresholded = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 2)

plt.figure(figsize=(12, 8), dpi=100)

# Plot the original image
plt.subplot(3, 1, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), cmap='gray')  # Overlay the segmented regions with jet colormap
plt.title('original image')
plt.axis('off')

# Plot the gray image
plt.subplot(3, 1, 2)
plt.imshow(gray, cmap='gray')  # Overlay the segmented regions with jet colormap
plt.title('gray image')
plt.axis('off')

# Plot the thresholded image
plt.subplot(3, 1, 3)
plt.imshow(thresholded, cmap='gray')  # Overlay the segmented regions with jet colormap
plt.title('thresholded image')
plt.axis('off')


plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
